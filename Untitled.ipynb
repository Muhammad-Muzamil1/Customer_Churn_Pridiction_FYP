```python
df=pd.read_csv(
"data_30000.csv"
)


df.dropna(
subset=["CustomerID"],
inplace=True
)



df=df[df["Quantity"]>0]



df["InvoiceDate"]=pd.to_datetime(
df["InvoiceDate"]
)



df["TotalAmount"]=(
df["Quantity"] *
df["UnitPrice"]
)



snapshot_date=(

df["InvoiceDate"].max()

+
pd.Timedelta(days=1)

)
```

```python
cust=df.groupby("CustomerID").agg({

"InvoiceDate":
lambda x:
(snapshot_date-x.max()).days,


"InvoiceNo":"nunique",


"TotalAmount":"sum",


"StockCode":"nunique"


}).reset_index()



cust.columns=[

"CustomerID",

"recency_days",

"frequency",

"monetary",

"unique_products"

]



cust.head()
```

```python
order_dates=df.groupby(
"CustomerID"
)["InvoiceDate"].apply(list)



def avg_gap(x):

    x=sorted(x)

    if len(x)<=1:
        return 0


    gaps=[]


    for i in range(1,len(x)):

        gaps.append(
            (x[i]-x[i-1]).days
        )


    return np.mean(gaps)



cust["mean_days_between_orders"]=(
cust["CustomerID"]
.map(
order_dates.apply(avg_gap)
)
)
```

```python
cust["avg_order_value"]=(
cust["monetary"]
/
cust["frequency"]
)
```

```python
cust["churn"]=(
cust["recency_days"]>90
).astype(int)
```

```python
cust["R"]=pd.qcut(

cust["recency_days"],

5,

labels=[5,4,3,2,1]

).astype(int)



cust["F"]=pd.qcut(

cust["frequency"].rank(method="first"),

5,

labels=[1,2,3,4,5]

).astype(int)



cust["M"]=pd.qcut(

cust["monetary"].rank(method="first"),

5,

labels=[1,2,3,4,5]

).astype(int)
```

```python
# =====================================
# Product Lookup
# =====================================


product_lookup = df[
    [
        "StockCode",
        "Description"
    ]
].drop_duplicates()



product_lookup.dropna(
    subset=["Description"],
    inplace=True
)



product_lookup.head()
```

```python
# =====================================
# Customer Product Purchase Matrix
# =====================================


purchase_matrix = df.pivot_table(

    index="CustomerID",

    columns="StockCode",

    values="Quantity",

    aggfunc="sum",

    fill_value=0

)



purchase_matrix.head()
```

```python
# =====================================
# Customer Similarity
# =====================================


from sklearn.metrics.pairwise import cosine_similarity



customer_similarity = cosine_similarity(

    purchase_matrix

)



similarity_df = pd.DataFrame(

    customer_similarity,

    index=purchase_matrix.index,

    columns=purchase_matrix.index

)


similarity_df.head()
```

```python
# =====================================
# Individual Customer Recommendation
# =====================================


def recommend_for_customer(
    customer_id,
    top_n=15
):


    if customer_id not in purchase_matrix.index:

        return []



    # products already purchased

    customer_products = purchase_matrix.loc[
        customer_id
    ]



    # find customers with similar interests

    similar_customers = (

        similarity_df[customer_id]

        .sort_values(
            ascending=False
        )

        .iloc[1:11]

    )



    recommendations = {}



    # collect products

    for similar_customer in similar_customers.index:


        bought_products = purchase_matrix.loc[
            similar_customer
        ]



        for product,quantity in bought_products.items():


            # only new products

            if customer_products[product]==0:


                recommendations[product] = (

                    recommendations.get(product,0)

                    +
                    quantity

                )



    # ranking

    recommended_products = sorted(

        recommendations.items(),

        key=lambda x:x[1],

        reverse=True

    )



    return [

        product

        for product,score in recommended_products[:top_n]

    ]
```

```python
# =====================================
# Churn Customer Recommendations
# =====================================


final_recommendations=[]



for customer in churned_customers.CustomerID:



    products = recommend_for_customer(

        customer,

        top_n=15

    )



    product_names=(

        product_lookup[

        product_lookup["StockCode"].isin(products)

        ]

        ["Description"]

        .tolist()

    )



    churn_probability = (

        churned_customers[

        churned_customers["CustomerID"]==customer

        ]

        ["churn_probability"]

        .values[0]

    )



    final_recommendations.append({

        "CustomerID":customer,


        "Churn_Probability":
        round(
            churn_probability,
            3
        ),


        "Recommended_Products":
        product_names

    })



recommendation_df=pd.DataFrame(
    final_recommendations
)



recommendation_df.head(10)
```

```python
# =====================================
# Coverage
# =====================================


total_churned=len(
    recommendation_df
)



customers_with_recommendations=(

recommendation_df[

recommendation_df["Recommended_Products"]

.apply(lambda x:len(x)>0)

]

.shape[0]

)



coverage=(

customers_with_recommendations /
total_churned

)*100



print(
"Recommendation Coverage:",
round(coverage,2),
"%"
)
```
